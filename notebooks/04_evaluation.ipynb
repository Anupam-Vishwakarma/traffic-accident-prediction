```python
# ======= Model Evaluation & Visualization =======

import pandas as pd
import matplotlib.pyplot as plt
import joblib
import numpy as np
import os
```

```python
# Create results directory if not present
os.makedirs("C:/Users/Anupam/Desktop/traffic-prediction-project/results", exist_ok=True)
```

```python
# Load metrics saved from Step 1
metrics = pd.read_csv("C:/Users/Anupam/Desktop/traffic-prediction-project/results/model_metrics.csv")
display(metrics)
```

```python
# =============================
# 1️⃣ Model Comparison Bar Chart
# =============================
plt.figure(figsize=(10,6))
metrics.plot(x='model', y=['accuracy','precision','recall','f1'], kind='bar', figsize=(10,6))
plt.title("Model Comparison")
plt.ylabel("Score")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("C:/Users/Anupam/Desktop/traffic-prediction-project/results/model_comparison.png")
plt.show()

print("✅ Saved model_comparison.png")
```

```python
# =============================
# 2️⃣ Feature Importance (XGBoost)
# =============================

xgb_model_path = "C:/Users/Anupam/Desktop/traffic-prediction-project/models/xgb.pkl"
xgb = joblib.load(xgb_model_path)

if hasattr(xgb, "feature_importances_"):
    fi = xgb.feature_importances_
    cols = pd.read_csv("C:/Users/Anupam/Desktop/traffic-prediction-project/data_processed/train_data.csv").drop('Severity', axis=1).columns
    imp = pd.Series(fi, index=cols).sort_values(ascending=False).head(20)

    plt.figure(figsize=(10,6))
    imp.plot(kind='bar')
    plt.title("XGBoost Feature Importance (Top 20)")
    plt.ylabel("Importance Score")
    plt.tight_layout()
    plt.savefig("C:/Users/Anupam/Desktop/traffic-prediction-project/results/feature_importance_xgb.png")
    plt.show()

    print("✅ Saved feature_importance_xgb.png")
else:
    print("⚠️ XGBoost model has no feature_importances_ attribute.")
```
